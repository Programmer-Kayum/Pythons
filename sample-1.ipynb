```python
#EDA

#distribution of case by sex
plt.figure(figsize=(12,6))
sns.countplot(data=data, x="sex", order= data['sex'].value_counts().index)
plt.title("Distribution of cases by sex")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(data=data, x="PermanentResidence", order= data['PermanentResidence'].value_counts().index)
plt.title("Distribution of cases by PermanentResidence")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(data=data, x="Religiousaffiliation", order= data['Religiousaffiliation'].value_counts().index)
plt.title("Distribution of cases by Religiousaffiliation")
plt.xticks(rotation=90)
plt.show()
```

```python
#Total cases over time

plt.figure(figsize=(8,4))
sns.lineplot(data=data, x ='averagemonthlyincome', y = 'Age', estimator = 'sum')
plt.title('Distrivution of Cases over time')
plt.show()
```
