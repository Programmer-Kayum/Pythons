```python
#                         Bar chart
```

```python
import numpy as np
import matplotlib.pyplot as plt
categories = ['Apple','Banana','Date','A','B','C']
values = [15,20,10,30,25,35]
plt.figure(figsize=(6,5))
plt.bar(categories,values,color='green')
plt.title("Sample Bar Chart")
plt.xlabel("categories")
plt.ylabel("Quantity")
plt.show()
```

```python
#                       Line Plot
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace( 0, 10, 100)
y = np.sin(x)
plt.figure(figsize=(6,5))
plt.plot(x,y)
plt.title("Sample Line Chart")
plt.xlabel("X")
plt.ylabel("Sin(x)")
plt.grid(True)
plt.show()
```

```python
# Histogram
```

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.random(100)
plt.figure(figsize=(6,5))
plt.hist(data, bins=50, alpha=.5, color = 'green')
plt.title("Sample Histogram Chart")
plt.xlabel("values")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter Plotting
```

```python
 import numpy as np
import matplotlib.pyplot as plt
x = np.random.random(100)
y = np.random.random(100)
plt.figure(figsize=(5,5))
plt.scatter(x,y)
plt.title("Sample Scatter Chart")
plt.xlabel("$X$")
plt.ylabel("$Y$")
plt.show()
```

```python
# PIE Chart Plotting
```

```python
import numpy as np
import matplotlib.pyplot as plt
labels = ['Red','Blue','Green','Yellow']
sizes = [213,130,245,210]
plt.figure(figsize=(6,5))
plt.pie(sizes,labels = labels,autopct='%1.1f%%',startangle=90)
plt.title("Sample PIE Chart")
plt.show()
```

```python
# Box Plot
```

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.figure(figsize=(6,5))
plt.boxplot(data,vert=True, patch_artist=True)
plt.title("Sample BOX Chart")
plt.xlabel("DATA")
plt.ylabel("Quantity")
plt.show()
```

```python
# Heatmap Plotting
```

```python
data = [np.random.rand(10,12)]
plt.figure(figsize=(6,5))
plt.imshow(data,cmap='hot',interpolation='nearest')
plt.show()
```
